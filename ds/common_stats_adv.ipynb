```python
for region, df_region in common.groupby('name'):
    df_region = df_region.sort_values('month__x')
    print('shape', df_region.shape)
    df_region = df_region.query('month__x >= 3')
    df_region = df_region.groupby('month__x').sum()
    print('shape df_region', df_region.shape)
    df_region = df_region[['Заражений за день', 'max_3']]
    df_region['max_3'] =  df_region['max_3']

    print(region)
    plt.figure()
    df_region.plot()
    plt.title(region);
    plt.legend();
    plt.show();
```

```python
import seaborn as sns

common_plot = common.copy()
common_plot.pop('month__y')
common_plot.pop('Unnamed: 0')
common_plot.pop('code')
common_plot.pop('Смертей за день')
common_plot.pop('Смертей')
common_plot.pop('Летальность, %')
common_plot.pop('year_')

plt.figure(figsize=(6, 6))

plt.title('Корреляционная матрица')
sns.heatmap(common_plot.rename(columns={
  'origin': 'Происхождение',
  'max_3': 'Максимальная цена',
  'min_3': 'Минимальная цена',
   'avg_3': 'Средняя цена',
  'month__x': 'Месяц',
  'Count_3_store': 'Кол-во на складе',
  'Count_3': 'Кол-во продаж',
   'is_sales': 'Данные продаж',
    'prod_pack_1_size': 'Размер упаковки'
}).corr(), 
            ax=plt.gca());
```
